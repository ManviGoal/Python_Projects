```python
###We can use below function to move legend in histplot. For other plots, commented command can help us to move legend location.
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
    
fig,ax=plt.subplots(figsize=(12,5))
sns.histplot(data=iris,x='sepal.length',y='petal.length',palette='coolwarm',hue='variety',ax=ax)
#ax.legend(iris['variety'].unique(),loc=3,bbox_to_anchor=[0.5,-.3,0,0])
move_legend(ax, "lower right")
```
